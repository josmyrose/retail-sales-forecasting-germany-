**Notebook 05 – Date Dimension & BI Preparation**

```python


import pandas as pd

# Load KPI dataset to infer date range
df = pd.read_csv(
    "/content/drive/MyDrive/retaildataanalysis/dataset/outputs/retail_kpi.csv"
)
df["date"] = pd.to_datetime(df["date"])

start_date = df["date"].min()
end_date = df["date"].max()

# Create monthly date range
dates = pd.date_range(start=start_date, end=end_date, freq="MS")

date_dim = pd.DataFrame({
    "date": dates,
    "year": dates.year,
    "quarter": dates.quarter,
    "month": dates.month,
    "month_name": dates.month_name(),
    "month_short": dates.strftime("%b"),
    "year_month": dates.strftime("%Y-%m"),
    "is_month_start": True,
    "is_month_end": True
})

# Validation
assert date_dim["date"].is_unique, "Date dimension has duplicate dates"

# Export
date_dim.to_csv(
    "/content/drive/MyDrive/retaildataanalysis/dataset/powerbi/date_dim.csv",
    index=False
)

print("Date dimension created for Power BI ✅")
```
